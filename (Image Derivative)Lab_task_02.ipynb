### Applying Different Derivates on Image

```python
from ImageDerivatives import ImageDerivatives
import matplotlib.pyplot as plt
```

```python
image_path = 'cameraman.tif'  
derivative_calculator = ImageDerivatives(image_path)
```

```python
print("Choose the derivative order to display:")
print("Press 0 for Zero-order derivative (Gaussian Smoothed Image)")
print("Press 1 for First-order derivative (Gradient Magnitude)")
print("Press 2 for Second-order derivative (Laplacian)")

choice = input("Enter your choice (0, 1, or 2): ")


options = {
    "0": ("Zero-order derivative", derivative_calculator.zero_order_derivative),
    "1": ("First-order derivative", derivative_calculator.first_order_derivative),
    "2": ("Second-order derivative", derivative_calculator.second_order_derivative)
          }

if choice in options:
    title, func = options[choice]
    result = func()
    plt.imshow(result, cmap="gray")
    plt.title(title)
    plt.axis("off")
    plt.show()
else:
    print("Invalid choice. Please enter 0, 1, or 2.")
```
